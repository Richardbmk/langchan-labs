```python
from langchain_core.prompts.chat import ChatPromptTemplate, MessagesPlaceholder
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

from langchain_openai import ChatOpenAI

from langchain_tavily import TavilySearch
from langchain_experimental.tools import PythonREPLTool

from langchain_classic.agents import create_tool_calling_agent, tool
from langchain_classic.agents import AgentExecutor

from langchain_classic.agents import AgentType, initialize_agent, load_tools

import os
```

```python
search = TavilySearchResults()
python = PythonREPLTool()
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant. Think step by step before responding."),
        ("placeholder", "{chat_history}"),
        ("human", "{input}"),
        ("placeholder", "{agent_scratchpad}")
    ]
)
```

```python
llm = ChatOpenAI()
```

```python
tools=[search, python]
```

```python
message_history = ChatMessageHistory()
```

```python
agent = create_tool_calling_agent(llm, tools, prompt)
```

```python
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
```

```python
agent1 = RunnableWithMessageHistory(
    agent_executor,
    lambda session_id: message_history,
    input_messages_key="input",
    history_messages_key="chat_history",
)
```

```python
response=agent1.invoke({"input": "When and where is Eurovision this 2026?"},config={"configurable": {"session_id": "session1"}})
print(response['output'])
```
