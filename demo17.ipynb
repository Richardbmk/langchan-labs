````python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_openai.chat_models import ChatOpenAI

from langchain_community.chat_message_histories import RedisChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

import os

model = ChatOpenAI()
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You're an assistant who's good at {ability}. Respond in 20 words or fewer",
        ),
        MessagesPlaceholder(variable_name="history"),
        ("human", "{input}"),
    ]
)
base_chain = prompt | model
````

````python
REDIS_URL = "redis://localhost:6379/0"
store = {}
````

````python
def get_message_history(session_id: str) -> RedisChatMessageHistory:
    return RedisChatMessageHistory(session_id, url=REDIS_URL)

redis_chain = RunnableWithMessageHistory(
    base_chain,
    get_message_history,
    input_messages_key="input",
    history_messages_key="history",
)
````

````python
redis_chain.invoke(
    {"ability": "math", 
     "input": "What does cosine mean?"}, 
    config={"configurable": {"session_id": "math-thread1"}},
)
````

````python
redis_chain.invoke(
    {"ability": "math", "input": "Tell me more!"},
    config={"configurable": {"session_id": "math-thread1"}},
)
````

````python
redis_chain.invoke(
    {"ability": "physics", 
     "input": "What is the theory of relativity?"}, 
    config={"configurable": {"session_id": "phy-thread1"}},
)
````

````python
redis_chain.invoke(
    {"ability": "physics", "input": "Tell me more!"},
    config={"configurable": {"session_id": "phy-thread1"}},
)
````

# ðŸ”§ Redis Container Setup & Debugging Guide

## ðŸ“¦ Step 1: Start Redis Container

Run the following command to start a Redis container with port mapping:

```bash
docker run -d --name redis-langchain -p 6379:6379 redis
```

**Flags explained:**
- `-d` â†’ Run in detached mode (background)
- `--name redis-langchain` â†’ Container name for easy reference
- `-p 6379:6379` â†’ Map Redis port from container to host machine

---

## ðŸ” Step 2: View Session History in Redis

### 2.1 Access the Container

Enter the running Redis container:

```bash
docker exec -it redis-langchain /bin/bash
```

### 2.2 Open Redis CLI

Once inside the container, start the Redis CLI:

```bash
redis-cli
```

### 2.3 List All Session Keys

View all stored session keys:

```bash
KEYS *
```

### 2.4 Inspect Session Messages

View the conversation history for a specific session:

```bash
LRANGE message_store:math-thread1 0 -1
```

> **ðŸ’¡ Tip:** Replace `math-thread1` with your actual session ID (e.g., `phy-thread1`)
