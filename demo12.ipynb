```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
prompt = ChatPromptTemplate.from_template("""
    You are a helpful assistant on {topic}. 
    Answer the following question: {question}
    """)
```

```python
llm = ChatOpenAI()
```

```python
output_parser = StrOutputParser()
```

```python
chain = prompt | llm | output_parser
```

```python
chain.invoke({"topic":"TV Shows","question":"Tell me about The Orange is the New Black TV show."})
```

```python
new_chain = RunnablePassthrough() | RunnablePassthrough.assign(topic=lambda x: "TV Shows") | prompt | llm | output_parser
```

```python
new_chain.invoke({"question":"Tell me about Inception"})
```
