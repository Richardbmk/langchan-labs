```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_community.document_loaders import WebBaseLoader
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
```

```python
URL1="https://medium.com/wardleymaps/on-being-lost-2ef5f05eb1ec"
URL2="https://medium.com/wardleymaps/finding-a-path-cdb1249078c0"
```

```python
loader = WebBaseLoader([URL1, URL2])
data = loader.load()
```

```python
prompt = ChatPromptTemplate.from_messages(
    [("system", "What are the five factors that can help me build a killer business strategy:\n\n{context}")]
)
```

```python
llm = ChatOpenAI(model="gpt-3.5-turbo")
```

```python
chain = create_stuff_documents_chain(llm, prompt)
```

```python
chain.invoke({"context": data})
```
